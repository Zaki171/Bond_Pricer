```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
par_val = 1000
coup_freq = 1
ytm = 0.05
T = 4
coups = [x for x in np.arange(0, 0.1, 0.01)]
durations = [duration(par_val, ytm, c, coup_freq, T) for c in coups]
print(durations)
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(coups, durations)
plt.show()
```

#### Increasing coupons increases duration ^^

```python
par_val = 1000
coup_freq = 1
ytms = [x for x in np.arange(0, 0.1, 0.01)]
T = 30
coup = 0.1
prices = [pricer(par_val, yld, coup, coup_freq, T) for yld in ytms]
```
